```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier


#Pandas reads file
music_data = pd.read_csv('music.csv')
music_data
```

```python
#input & output data
X = music_data.drop('genre', axis=1)
X
```

```python
y = music_data.drop(columns=['age', 'gender'], axis=1)
y
```

```python
#Build, fit & evaluate model
model = DecisionTreeClassifier()
model.fit(X, y)

pred = model.predict([ [21, 0], [30, 1] ])
pred
```
