```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

#Pandas read file
app_det = pd.read_csv('Applicant-details.csv')
app_det
```

```python
X = app_det.drop(columns=['Marital_Status', 'House_Ownership', 'Vehicle_Ownership(car)', 'Occupation', 'Residence_City', 'Residence_State'], axis=1)
X
```

```python
y = app_det.drop(columns=['Applicant_ID', 'Annual_Income', 'Applicant_Age', 'Work_Experience', 'Years_in_Current_Employment', 'Years_in_Current_Residence', 'Loan_Default_Risk'], axis=1)
y
```

```python
#Build, fit & predict model
model = DecisionTreeClassifier()
model.fit(X, y)

predictions = model.predict([ [19865, 1509721, 66, 8, 4, 12, 0] ])
predictions
```
