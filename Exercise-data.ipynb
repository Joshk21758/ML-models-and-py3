```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

#Read file
injury_data = pd.read_csv('injury_data.csv')
injury_data
```

```python
#Input & output data
X = injury_data.drop(columns=['Recovery_Time', 'Likelihood_of_Injury'])
X
```

```python
y = injury_data[['Recovery_Time', 'Likelihood_of_Injury']]
y
```

```python
#Build, fit & evaluate model
model = DecisionTreeClassifier()
model.fit(X, y)

pred = model.predict([ [24, 66.251933, 175.732429, 1, 0.457929] ])
pred
```
