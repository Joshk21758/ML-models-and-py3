```python
import pandas as pd
import numpy as np

#Read file
heart_data = pd.read_csv('Training.csv')
heart_data
```

```python
heart_data.head()
```

```python
#Input & Output data
X = heart_data.drop(columns=['Outcome'])
X
```

```python
y = heart_data['Outcome']
y
```

```python
from sklearn.model_selection import train_test_split
#Train test data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
X_train.shape, y_train.shape
X_test.shape, y_test.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(heart_data['Outcome'])
plt.xlabel('Actual values')
plt.ylabel('Predicted values')
plt.title('Heart disease(Positive/Neg)')
plt.show()
```

```python
heart_data['Outcome'].value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
#Build, fit & evaluate model
model = RandomForestClassifier(
    criterion='gini',
    max_depth=10,
    min_samples_split=8,
    random_state=5,
    
)
model.fit(X_train, y_train)
pred = model.predict(X_test)
pred
```

```python
from sklearn.metrics import accuracy_score
#Evaluate model
score = accuracy_score(y_test, pred)
print("Accuracy score: ", score)
```
