```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier

#Pandas reads file
iris_data = pd.read_csv('iris.csv')
iris_data
```

```python
#Input & output data
X = iris_data.drop('species', axis=1)
X
```

```python
y = iris_data['species']
y
```

```python
#Build, fit & predict model
model = DecisionTreeClassifier()
model.fit(X, y)

predictions = model.predict([ [4.9, 3.0, 1.4, 0.2] ])
predictions
```
